```python
import matplotlib.pyplot as plt
step = [0,1000,2000,3000,4000,5000]
randinit = [0.134, 0.292, 0.332, 0.350, 0.378, 0.378]
mocopretrain = [0.146, 0.468, 0.491, 0.497, 0.513, 0.650]
unetrrandominit = [0.188, 0.330, 0.374, 0.388, 0.428, 0.438]

# Using a seaborn style for a more sophisticated look
plt.style.use('classic')
plt.figure(figsize=(12, 7))
plt.plot(step, randinit, label='randinit', marker='o', linestyle='-', color='blue')
plt.plot(step, mocopretrain, label='mocopretrain', marker='s', linestyle='--', color='green')
plt.plot(step, unetrrandominit, label='unetrrandominit', marker='^', linestyle='-.', color='red')
plt.title('Different models IoU scores', fontsize=16)
plt.xlabel('Epochs', fontsize=14)
plt.ylabel('IoU Values', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.xticks(step, fontsize=12)
plt.yticks(fontsize=12)



# Show the plot
plt.show()
```
